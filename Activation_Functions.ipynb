# **ACTIVATION FUNCTIONS**

> **Activation functions are mathematical equations that determine the output of a neural network model.**

1.  **`Hyperbolic Tangent   [ -1 , 1 ]`**

m_tanh.jpg    
2.   **`Rectified [ 0 , 1 ]`**

relu.jpg
















```python
import math
```

```python
def tanh (x):
  return ((1-math.exp(-2*x))-1 + math.exp(-2*x))
def relu (x):
  return (max(0,x))  
```

```python
print (tanh(-9))
print (tanh(4))
```

```python
print (relu(-7))
print (relu(10))
```
