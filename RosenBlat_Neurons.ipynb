# RosenBlat Neurons

1.   **These neurons have the ability to adjust weights , thereby finding the appropriate combination of inputs and weights.**
2.   **Neuron can be modified to implement multiple Boolean functions in one code**


> Rosenblattperceptron.png





  




```python
import numpy as np
from numpy import array, dot, random
from random import choice
```

**Step Function**

```python
step_function = lambda x: 0 if x<40 else 1
training_data= [
                (array([0,0,1]),0),
                (array([0,1,1]),0),
                (array([1,0,1]),0),
                (array([1,1,1]),0)

]
```

```python
w = random.rand(3)
bias= .1
errors =[]
eta= 0.1
n = 10000
```

**Adjusting weights by finding errors in the predicted values**

```python
 for i in range(n):
   x,expected =choice(training_data)
   result= dot(w,x)
   e=expected-step_function(result)
   errors.append(e)
   w +=eta*e*x
```

**Finally checking for accuracy**

```python
for x,_ in (training_data):
  result=dot(x,w)
  print ("{} :{} -> {}".format(x[0:3],result,step_function(result)))
```
